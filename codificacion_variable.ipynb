```python
import pandas as pd
import re
import unidecode
```

```python
# Cargamos el dataframe
df = pd.read_excel("Historico de Siniestros.xlsx")
```

```python
# Vemos el dataframe
df.head()
```

```python
df["DE_LUGAR_OCURRENCIA"] = df["DE_LUGAR_OCURRENCIA"].str.lower()
```

```python
import unidecode

def remover_acentos(x):
    """Creamos la funcion para remover los acentos"""
    if isinstance(x, str):
        return unidecode.unidecode(x)
    else:
        return x

# Aplicar la función a la columna "DE_LUGAR_OCURRENCIA"
df["DE_LUGAR_OCURRENCIA"] = df["DE_LUGAR_OCURRENCIA"].apply(remover_acentos)
```

```python


def remove_special_characters(text, remove_digits=False):
    """Funcion para remover los caracteres especiales"""
    pattern = r'[^a-zA-Z0-9\s]' if not remove_digits else r'[^a-zA-Z\s]'
    text = re.sub(pattern, '', text)
    return text

# Asegurarse de que todos los valores en la columna sean cadenas
df["DE_LUGAR_OCURRENCIA"] = df["DE_LUGAR_OCURRENCIA"].astype(str)

# Aplicar la función a la columna "DE_LUGAR_OCURRENCIA"
df["DE_LUGAR_OCURRENCIA"] = df["DE_LUGAR_OCURRENCIA"].apply(remove_special_characters)
```

```python
def categorizar_lugar(lugar):
    """Creamos una columna, para identificar donde fue el sunuestro"""
    if "avenida" in lugar.lower() or "av" in lugar.lower():
        return "Avenida"
    elif "calle" in lugar.lower():
        return "Calle"
    elif "autopista" in lugar.lower() or "aut" in lugar.lower():
        return "Autopista"
    else:
        return "Otro"

df["Categoria_Lugar"] = df["DE_LUGAR_OCURRENCIA"].apply(categorizar_lugar)
df["Categoria_Lugar"].value_counts()
```
